```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return 4*x + 0.5*y
    



def euler_method(f, y0, x0, x_end, h):
    n = int((x_end - x0) / h)
    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        y0 += h * f(x0, y0) 
        x0 += h
        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values


def rk2_method(f, y0, x0, x_end, h):
    n = int((x_end - x0) / h)
    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h, y0 + k1)
        y0 += (k1 + k2) / 2
        x0 += h
        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values


def rk4_method(f, y0, x0, x_end, h):
    n = int((x_end - x0) / h)
    x_values = [x0]
    y_values = [y0]

    for i in range(n):
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h / 2, y0 + k1 / 2)
        k3 = h * f(x0 + h / 2, y0 + k2 / 2)
        k4 = h * f(x0 + h, y0 + k3)
        y0 += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        x0 += h
        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values

x0 = 0
y0 = 1  
x_end = 2
h = 0.1 

x_values_euler, y_values_euler = euler_method(f, y0, x0, x_end, h)
x_values_rk2, y_values_rk2 = rk2_method(f, y0, x0, x_end, h)
x_values_rk4, y_values_rk4 = rk4_method(f, y0, x0, x_end, h)

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.plot(x_values_euler, y_values_euler, label='Euler Method', color='blue')
plt.title("Euler's Method")
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x_values_rk2, y_values_rk2, label='RK2 Method', color='orange')
plt.title("RK2 Method")
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x_values_rk4, y_values_rk4, label='RK4 Method', color='green')
plt.title("RK4 Method")
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x_values_euler, y_values_euler, label='Euler Method', color='blue')
plt.plot(x_values_rk2, y_values_rk2, label='RK2 Method', color='orange')
plt.plot(x_values_rk4, y_values_rk4, label='RK4 Method', color='green')
plt.title('Comparison of Euler, RK2, and RK4 Methods')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
